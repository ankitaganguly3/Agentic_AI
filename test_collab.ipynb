```python
student_info = {
    1001: {"name": "Alice", "age": 20, "grade": "A"},
    1002: {"name": "Bob", "age": 22, "grade": "B"},
    1003: {"name": "Charlie", "age": 21, "grade": "A-"}
}

for student_id, student_info in student_info.items():
    print(f"Student ID: {student_id}")
    for key, value in student_info.items():
        print(f"  {key.capitalize()}: {value}") 
    print("_"*40)          
```
